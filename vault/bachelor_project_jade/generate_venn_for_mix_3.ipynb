```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib_venn import venn2
```

```python
df = pd.read_csv('plate_3_targets_detected.csv')
```

```python
positive = df['detected_pos'] == True
negative = df['detected_neg'] == True
```

```python
only_positive = sum(positive & ~negative)
only_negative = sum(~positive & negative)
both = sum(positive & negative)
none = sum(~positive & ~negative)
```

```python
plt.figure(figsize=(8,6))
venn = venn2(subsets=(only_positive, only_negative, both), 
             set_labels=('Positive Mode', 'Negative Mode'))
```

```python
plt.title(f'Detection Venn Diagram\n({len(df)} mol√©cules total)')
plt.show()
```
